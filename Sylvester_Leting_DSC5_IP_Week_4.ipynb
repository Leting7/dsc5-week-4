```
#changed the keys to almost be similar to the statistical denotations
mu = 12.437500
sigma = 4.289036
α = 0.05
# we look up the critical z value that corresponds to this alpha level
z_critical = 1.96
x_bar = 12.548600
N = 26
SE = sigma/np.sqrt(N)
z_stat = (x_bar - mu)/SE
print(z_stat)
from scipy.stats import norm
import scipy.stats as stats
prob = stats.norm.cdf(z_stat)
print(prob)
# interpretation of the p value
if prob < 0.05:    # alpha value is 0.05 or 5%
   print(" we rejecting null hypothesis")
else:
  print("we fail to reject null hypothesis")
```

with the p value score seen above the level of significance we ascertain well that we shall retain the null hypothesis
